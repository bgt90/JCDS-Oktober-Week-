# Content Based Filtering With Anime Dataset
## 1 User

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df_anime= pd.read_csv('anime.csv', usecols=['anime_id','name','rating','genre']).dropna(subset=['genre']).reset_index(drop=True)
```

```python
df_anime
```

```python
cvr=CountVectorizer(tokenizer=lambda x:x.split(', '))
df_genre=pd.DataFrame(cvr.fit_transform(df_anime['genre']).toarray(), columns=cvr.get_feature_names_out())
df_genre.head()
```

```python
df_genre=pd.concat([df_anime[['anime_id','name','rating']],df_genre], axis=1)
df_genre.head()
```

```python
# df dengan anime yang disukai user
df_liked=df_genre.loc[df_genre['name'].isin(['Naruto', 'One Piece','Detective Conan'])].copy()
df_liked
```

```python
for i in cvr.get_feature_names_out(): #nama-nama genre
    df_liked[i]=df_liked['rating']*df_liked[i]
    
df_liked
```

```python
user_feature_vector=df_liked[cvr.get_feature_names_out()].sum()/df_liked[cvr.get_feature_names_out()].sum().sum()
# user_feature_vector
```

```python
df_rec=df_genre.loc[~df_genre['name'].isin(['Naruto', 'One Piece','Detective Conan'])].drop(columns='rating').reset_index(drop=True)
```

```python
for i in cvr.get_feature_names_out():
    df_rec[i]= df_rec[i]*user_feature_vector[i]
    
df_rec.head()
```

```python
df_rec['pred_score']=df_rec.drop(columns='anime_id').sum(axis=1)
```

```python
df_rec[['name','anime_id','pred_score']].sort_values('pred_score',ascending=False).head(30)
```

## With Multiple

```python
df_user_item=pd.DataFrame({
    'user':['user 1','user 2','user 3','user 4'],
    32281:[7,8,9,8],
    431:[9,8,9,7],
    1698:[8,8,8,7],
    4722:[9,7,10,8]
})
df_user_item
```

```python
df_genre
```

```python
df_item_feature=df_genre.loc[df_genre['anime_id'].isin(df_user_item.columns[1:])].drop(columns='rating')
df_item_feature
```

```python
user_item=np.array(df_user_item.drop(columns='user'))
item_feature=np.array(df_item_feature.drop(columns=['anime_id','name']))

n_user=user_item.shape[0]
n_item= user_item.shape[1]
n_feature=item_feature.shape[1]

user_feature=np.empty((n_user,n_feature))

for i in range(n_user):
    user_feature_vector=np.matmul(user_item[i,:], item_feature)
    user_feature_vector=user_feature_vector/user_feature_vector.sum()
    user_feature[i,:]= user_feature_vector
```

```python
df_user_feature=pd.DataFrame(user_feature, index=df_user_item['user'], columns=df_item_feature.columns[2:])
```

```python
df_item_feature_new=df_genre.loc[~df_genre['anime_id'].isin(df_user_item.columns[1:])].drop(columns='rating').reset_index(drop=True)
```

```python
item_feature_new= np.array(df_item_feature_new.drop(columns=['anime_id','name']))
# item_feature_new
n_item_new= item_feature_new.shape[0]
# n_item_new
user_item_score_new=np.empty((n_user,n_item_new))

for i in range(n_user):
    user_item_score_=np.matmul(item_feature_new, user_feature[i,:])
    user_item_score_new[i,:]=user_item_score_
```

```python
df_score=pd.DataFrame(user_item_score_new).T
```

```python
df_score.columns=df_user_item['user']
df_score
```

```python
df_score=pd.concat([df_score,df_item_feature_new[['anime_id','name']]],axis=1)
```

# Collaborative Filtering: Model based

```python
import pandas as pd
import numpy as sns

import seaborn as sns

from surprise import Reader,Dataset, SVD, BaselineOnly, accuracy
from surprise.model_selection import cross_validate, train_test_split, GridSearchCV
```

```python
df=pd.read_csv('u.data', sep='\t', names=['user_id','item_id','rating','timestamp'])
```

## Data

```python
df
```

```python
df['rating'].value_counts()
```

```python
df.pivot_table(values='rating',index='user_id',columns='item_id') # user item matrix
```

```python
reader=Reader(rating_scale=(0,5))
data= Dataset.load_from_df(df[['user_id','item_id','rating']],reader)
```

```python
data.df.head()
```

## Validation

```python
# data splitting
trainset, testset= train_test_split(data,test_size=0.25, random_state=101)
```

```python
#SVD
algo=SVD()
algo.fit(trainset)
prediction=algo.test(testset)
```

```python
# prediction
```

```python
accuracy.rmse(prediction)
```

```python
# ALS
bsl_option= {'method':'als',
            'n_epoch':5,
            'reg_u':12,
            'reg_i':5}
algo=BaselineOnly(bsl_options=bsl_option)
algo.fit(trainset)
prediction= algo.test(testset)
```

```python
accuracy.rmse(prediction)
```

## Cross Validasi

```python
# SVD
algo=SVD()
cv_svd= cross_validate(algo,data,measures=['RMSE','MAE'], cv=5, verbose=True)
```

```python
print('rmse cv mean', cv_svd['test_rmse'].mean())
```

```python
# ALS
bsl_option={
    'method':'als',
    'n_epoch':5,
    'reg_u':12,
    'reg_i':5
}
algo= BaselineOnly(bsl_options=bsl_option)
cv_als=cross_validate(algo,data, measures=['RMSE','MAE'], cv=5, verbose=False)
```

```python
print('rmse cv mean', cv_als['test_rmse'].mean())
```

berdasarkan hasil cross validasi, SVD memiliki performa sedikit lebih baik dibanding ALS

## Hyperparameter Tunning

```python
param_grid={
    'n_epochs':[5,10,20],
    'lr_all':[0.002,0.005],
    'reg_all':[0.2,0.4,0.6]
}
# n_epochs – The number of iteration of the SGD procedure. Default is 20.
# lr_all – The learning rate for all parameters. Default is 0.005
# reg_all – The regularization term for all parameters. Default is 0.02
# https://surprise.readthedocs.io/en/stable/matrix_factorization.html
```

```python
gs= GridSearchCV(SVD, param_grid, measures=['RMSE','MAE'],cv=5)
gs.fit(data)
```

```python
print(gs.best_score['rmse'])
print(gs.best_params['rmse'])
```

```python
print(gs.best_score['mae'])
print(gs.best_params['mae'])
```

```python
# performace to data test
algo= SVD(n_epochs= 20, lr_all= 0.005, reg_all= 0.2)
algo.fit(trainset)
prediction= algo.test(testset)
accuracy.rmse(prediction)
```

```python
# before after tunning
# before

algo= SVD()
algo.fit(trainset)
prediction= algo.test(testset)
accuracy.rmse(prediction)
```

before tunning memiliki performa sedikit lebih baik
